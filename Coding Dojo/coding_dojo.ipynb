```python
print("Jamesson Leandro")
print("Rodrigo Coelho")
from random import randint, uniform
```

```python
# Caracteristicas Jogador
vida_player = 100
armadura_magica_player = 10
armadura_fisica_player = 10
ataque_player = 10
pocoes_vida_player = 1
ganho_pocao = 25

# Caracteristicas Inimigo
nome_inimigo = 'Covid'
vida_inimigo = 100
armadura_magica_inimigo = 10
armadura_fisica_inimigo = 10
ataque_inimigo = 10
pocoes_vida_inimigo = 1

contador_rodada = 1

# Funções de ataque
def funcao_ataque():
    return randint(5,10)

def funcao_magia():
    x = uniform(0,1)
    
    if x >= 0 and x < 0.1:
        return -10
    elif x >= 0.1 and x < 0.3:
        return 0
    elif x >= 0.3 and x < 0.8:
        return 10
    else:
        return 30

nome = input('Qual seu nome? ')
print(f'Ola {nome}, sua armadura mágica de : ', armadura_magica_player)
print(f'Ola {nome}, sua armadura física de :', armadura_fisica_player)

# Loop principal
while vida_player > 0:
    
    # Tabela de informações para o payer
    print('RODADA ', contador_rodada)
    print('Vida: ', vida_player)

    print('Vida inimigo: ', vida_inimigo)
    
    escolha = input('Qual seu movimento? (1)Atacar (2)Atacar com magia (3) Usar pocao: ')
    
    if escolha == '3' and pocoes_vida_player == 0:
        escolha = input('Voce nao tem mais pocoes! Escolha apenas (1) Ataque ou (2) Ataque com magia: ')      
        
    # Ataque sem magia                             
    if escolha == '1':
        ataque = funcao_ataque()
        vida_inimigo -= ataque * (1 - armadura_fisica_inimigo/100)
        print('Ataque bem sucedido! Vida do Inimigo: ', vida_inimigo)
        
    # Ataque com magia
    elif escolha == '2':
        ataque = funcao_magia()
        if ataque < 0:
            vida_player += ataque
            print('Infelizmente seu ataque se voltou contra você! Você perdeu: ', ataque)
            if vida_player <= 0:
                print('Você morreu com seu próprio ataque')
        
        elif ataque == 0:
            print('Seu ataque não funcionou')
            
        else:    
            vida_inimigo -= ataque * (1 - armadura_magica_inimigo/100)
            print('Seu dano com magia foi de: ', ataque)
        print('Vida do Inimigo: ', vida_inimigo)
    
    # uso de poção pelo player
    else:
        vida_player += ganho_pocao
        pocoes_vida_player -= 1
        
    print('')
    print('Turno do inimigo!')
    
    # Verifica se inimigo tem pocoes de vida
    
    if pocoes_vida_inimigo > 0:
        escolha_inimigo = randint(0,2)
    else:
        escolha_inimigo = randint(0,3)
    
    # Acao da escolha do inimigo
    
    if escolha_inimigo == 0:
        ataque = funcao_ataque()
        vida_player -= ataque * (1 - armadura_fisica_player/100)
        print('Seu inimigo te atacou. Sua vida restante: ', vida_player)
    
    # Ataque mágico do inimigo
    elif escolha_inimigo == 1:
        ataque = funcao_magia()
        if ataque < 0:
            vida_inimigo += ataque
            print('Felizmente, o ataque de seu inimigo se voltou contra ele')
            if vida_inimigo <= 0:
                print('Seu inimigo morreu. Você ganhou o jogo!')
                break
        
        elif ataque == 0:
            print('Seu inimigo tentou atacar, mas nada aconteceu!')
            
        else:    
            vida_player -= ataque * (1 - armadura_magica_player/100)
            print('Seu inimigo usou magia contra você. Você tomou um dano de: ', ataque)
        print('Sua vida agora: ', vida_player)
        
    if vida_inimigo <= 0:
        print('Parabens!! Você ganhou!!')
        break
    
    print('')
    contador_rodada += 1
```
